Problem Statement 

Input :  A set of Activities with their start and finish time. 

Output : Maximum no. of non-conflicting activities that may be scheduled.  

```python
#List of random activities with their start and finish time. 
activity_lst = [(12,20),(1,10),(2,9),(3,8),(4,7),(6,9),(10,15),(0,21),(5,6),(6,10)]
```

```python
# Sorting the activities according to their finish time 

sort_act_lst = sorted(activity_lst, key=lambda x: x[1])
print(sort_act_lst) 
```

```python
# First Finish Time Approach
def ftf_act_scheduler(act_lst):
    """
    Schedule activities based on finish times.

    Parameters:
    - act_lst (list): List of activities represented by tuples (start_time, finish_time).

    Returns:
    - list: Scheduled activities without overlaps, each chosen based on earliest finish time.
    """
    result = [] 
    j = 0  
    
    result.append(act_lst[j])

    for i in range(1, len(act_lst)):
        if act_lst[j][1] <= act_lst[i][0]:
            j = i
            result.append(act_lst[i])

    return result
```

```python
#function call to first finish time activity schedular
result_ftf = ftf_act_scheduler(sort_act_lst)
print(len(result_ftf)," Activities scheduled  via First Finish Time--> " , result_ftf) 
```

```python
# Latest Start Time Approach
def lst_act_scheduler(act_lst):
    """
    Schedule activities based on latest start times.

    Parameters:
    - act_lst (list): List of activities represented by tuples (start_time, finish_time).

    Returns:
    - list: Scheduled activities without overlaps, each chosen based on latest start time.
    """
    result = []  
    j = 0
 
    result.append(act_lst[j])

    for i in range(1, len(act_lst)):
        if act_lst[j][0] >= act_lst[i][1]:
            j = i
            result.append(act_lst[i])

    return result[::-1]
```

```python
# Sorting the activities according to their latest start time 
desc_sort_act_lst = sorted(activity_lst, key=lambda x: x[0], reverse=True)

print(desc_sort_act_lst)
#function call to latest start time activity schedular

result_lst = lst_act_scheduler(desc_sort_act_lst)
print(len(result_lst)," Activities scheduled via Latest Start Time --> " , result_lst) 
```

```python
import matplotlib.pyplot as plt
from itertools import cycle

activities = activity_lst

fig, ax = plt.subplots()

# Cycle through colors for each activity
colors = cycle(['b', 'g', 'r', 'c', 'm', 'y'])
for index, (start, finish) in enumerate(activities):
    ax.plot([start, finish], [index, index], 'o-', markersize=10, color=next(colors), label=f'Act {index + 1}')

# Set labels and title
ax.set(xlabel='Time', ylabel='', title='Activity Timeline Before Scheduling')
ax.legend()

plt.show()
```
